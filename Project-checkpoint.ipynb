```python
IMAGE_HEIGHT, IMAGE_WIDTH = 64, 64

SEQUENCE_LENGTH = 20

DATASET_DIR = 'UCF50'

CLASSES_LIST = ["TaiChi", "Punch", "Nunchucks", "Fencing"]
```

```python
def frames_extraction(video_path):
  '''
  This function will extract required frames from the video after resizing and normalizing them.
  Args:
    video_path: The path of the  video in the disk, whose frames are to be extracted.
  Returns:
    frames_list: A list containing the resized  and normalized frame of the video
  '''

  #Declaring frame list
  frame_list = []

  #Read the video using Video Capture object
  video_read = cv2.VideoCapture(video_path)

  #Counts the frmaes in the video
  vid_frame_count = int(video_read.get(cv2.CAP_PROP_FRAME_COUNT))

  #Calculates the interval after which frames will be added to the list
  sk_frame_count = max(int(vid_frame_count/SEQUENCE_LENGTH), 1)

  for frame_count in range(SEQUENCE_LENGTH):

    #Set current frame position of the video
    video_read.set(cv2.CAP_PROP_POS_FRAMES, frame_count * sk_frame_count)

    #Reading frame from the video
    success, frame = video_read.read()

    #Check if the video u=is not successfuly read
    if not success:
      break

    #Resize the frame
    resized_frame = cv2.resize(frame, (IMAGE_HEIGHT, IMAGE_WIDTH))

    #Normalize the frame by dividing it with 255 so that each pixel calue lies between 0 and 1
    normalized_frame = resized_frame/255

    #append the normalized frame to the frame list
    frame_list.append(normalized_frame)

  #Release the VidoeCapture Object
  video_read.release()

  return frame_list
```

```python
def create_dataset():
  '''
  This function will extract the data from the selected class and create the required dataset
  Returns:
    features:         A list containing the extracted frames of the videos
    labels:           A list containing the indexes of the classes associated the videos
    vidoe_file_path:  A list containing the paths of the videos in the disk.
  '''

  #Createing the empty list
  features = []
  labels = []
  video_file_path = []

  for class_index, clas_name in enumerate(CLASSES_LIST):

    print(f'Extracting data of class: {clas_name}')

    #Get list of video files present in the specific class name directory.
    files_list = os.listdir(os.path.join(DATASET_DIR, clas_name))

    for file_name in files_list:

      #Get full vidoe path
      video_file_paths = os.path.join(DATASET_DIR, clas_name, file_name)

      #Exctract the frames from the video file
      frames = frames_extraction(video_file_paths)

      #Check if the extracted frames is equl to Sequece length
      #Ignore the video having frames less than Sequence Length
      if len(frames) == SEQUENCE_LENGTH:

        #Append the data according to the list created
        features.append(frames)
        labels.append(class_index)
        video_file_path.append(video_file_paths)

  features = np.asarray(features)
  labels = np.array(labels)

  return features, labels, video_file_path
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, TimeDistributed, LSTM, Flatten, Dense
from sklearn.model_selection import train_test_split

# Reshape for CNN+LSTM: (samples, timesteps, height, width, channels)
X_cnn_lstm = features  # Already in (521, 20, 64, 64, 3)

# Repeat labels for sequence-based learning (each video has 1 label)
y_cnn_lstm = labels  # No need to repeat since LSTM works per video sequence

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_cnn_lstm, y_cnn_lstm, test_size=0.25, random_state=42, shuffle=True
)

# Define CNN + LSTM model
cnn_lstm_model = Sequential([
    # Apply CNN to each frame using TimeDistributed
    TimeDistributed(Conv2D(32, (3, 3), activation='relu', padding='same'), input_shape=(20, 64, 64, 3)),
    TimeDistributed(MaxPooling2D(pool_size=(2, 2))),
    TimeDistributed(Conv2D(64, (3, 3), activation='relu', padding='same')),
    TimeDistributed(MaxPooling2D(pool_size=(2, 2))),
    TimeDistributed(Flatten()),  # Flatten feature maps per frame

    # LSTM layer for temporal learning
    LSTM(64, return_sequences=False),  # Output single vector per sequence
    Dense(128, activation='relu'),
    Dense(len(set(labels)), activation='softmax')  # Final output layer
])

# Compile model
cnn_lstm_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train model
cnn_lstm_model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))

# Evaluate model
cnn_lstm_accuracy = cnn_lstm_model.evaluate(X_test, y_test)[1]
print(f"CNN + LSTM Accuracy: {cnn_lstm_accuracy:.2f}")
```
